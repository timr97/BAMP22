```python
# utility methods to identify special terms and characters
def give_emoji_free_text(text):
    """
    Removes emoji's from tweets
    Accepts:
        Text (tweets)
    Returns:
        Text (emoji free tweets)
    """
    emoji_list = [c for c in text if c in emoji.UNICODE_EMOJI['en']]
    clean_text = ' '.join([str for str in text.split() if not any(i in str for i in emoji_list)])
    return clean_text

def url_free_text(text):
    '''
    Cleans text from urls
    '''
    text = re.sub(r'http\S+', '', text)
    return text

def at_free_text(text):
    # \S*@\S*\s?
    text = re.sub(r'\S*@\S*\s?', '', text)
    return text
```

```python
# Apply the function above and get tweets free of emoji's
call_emoji_free = lambda x: give_emoji_free_text(x)

# Apply `call_emoji_free` which calls the function to remove all emoji's
df['emoji_free_text'] = df['text'].apply(call_emoji_free)

# Create a new column with url free tweets
df['at_free_text'] = df['emoji_free_text'].apply(at_free_text)

# Create a new column with url free tweets
df['url_free_text'] = df['at_free_text'].apply(url_free_text)
```

```python
# Tokenizer
tokenizer = Tokenizer(nlp.vocab)


# Custom stopwords
custom_stopwords = ['hi','\n','\n\n', '&amp;', ' ', '.', '-', 'got', "it's", 'it’s', "i'm", 'i’m', 'im', 'want', 'like', '$', '@', '+']

# Customize stop words by adding to the default list
STOP_WORDS = nlp.Defaults.stop_words.union(custom_stopwords)

# ALL_STOP_WORDS = spacy + gensim + wordcloud
ALL_STOP_WORDS = STOP_WORDS.union(SW).union(stopwords)

tokens = []

for doc in tokenizer.pipe(df['url_free_text'], batch_size=500):
    doc_tokens = []    
    for token in doc: 
        if token.text.lower() not in STOP_WORDS:
            doc_tokens.append(token.text.lower())   
    tokens.append(doc_tokens)

# Makes tokens column
df['tokens'] = tokens
```

```python
# Utility method to do lemmatization
def get_lemmas(text):
    '''Used to lemmatize the processed tweets'''
    lemmas = []
    
    doc = nlp(text)
    
    # Something goes here :P
    for token in doc: 
        if ((token.is_stop == False) and (token.is_punct == False)) and (token.pos_ != 'PRON'):
            lemmas.append(token.lemma_)
    
    return lemmas
```

```python
# Make tokens a string again
df['tokens_back_to_text'] = [' '.join(map(str, l)) for l in df['tokens']]

df['lemmas'] = df['tokens_back_to_text'].apply(get_lemmas)
```

```python
# Tokenizer function
def tokenize(text):
    """
    Parses a string into a list of semantic units (words)
    Args:
        text (str): The string that the function will tokenize.
    Returns:
        list: tokens parsed out
    """
    # Removing url's
    pattern = r"http\S+"
    
    text = re.sub(pattern, "", text) # https://www.youtube.com/watch?v=O2onA4r5UaY
    text = re.sub('[^a-zA-Z 0-9]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # Remove punctuation
    text = re.sub('\w*\d\w*', '', text) # Remove words containing numbers
    text = re.sub('@*!*\$*', '', text) # Remove @ ! $
    text = re.sub('\d+', '', text) # get rid of any number

    tokens = text.lower().split() # Make text lowercase and split it
    # remove tokens who have only one or two characters
    tokens = [t for t in tokens if len(t) > 2]

    return tokens
```

```python
# Make lemmas a string again
df['lemmas_back_to_text'] = [' '.join(map(str, l)) for l in df['lemmas']]

# Apply tokenizer
df['lemma_tokens'] = df['lemmas_back_to_text'].apply(tokenize)
```

```python
# Defining a function to loop over number of topics to be used to find an 
# optimal number of tipics
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics
    and u_mass coherence

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the 
    LDA model with respective number of topics
    """
    coherence_c_v_topic = []
    coherence_u_mass_topic = []
    model_list_topic = []
    log_list = []
    for num_topics in range(start, limit, step):
        log_dic = {}
        print(f'Log: Run with {num_topics} topics.')
        log_dic['topics'] = num_topics
        model = LdaMulticore(corpus=corpus, num_topics=num_topics, id2word=dictionary, random_state=267, passes=10) # passes=5, --> maybe test this. I think default is 1
        model_list_topic.append(model)

        # c v
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        cm_score = coherencemodel.get_coherence()
        print(f'Log: Run ends with a c_v score of {cm_score}.')
        log_dic['c_v'] = cm_score
        coherence_c_v_topic.append(cm_score)

        # u mass
        coherencemodel = CoherenceModel(model=model, corpus=corpus, coherence='u_mass')
        cm_score = coherencemodel.get_coherence()
        print(f'Log: Run ends with a u_mass score of {cm_score}.')
        log_dic['u_mass'] = cm_score
        coherence_u_mass_topic.append(cm_score)
        log_list.append(log_dic)

    return model_list_topic, coherence_c_v_topic, coherence_u_mass_topic, log_list
```

```python
# Defining a function to loop over number of topics to be used to find an 
# optimal number of tipics
def advanced_compute_coherence_values(dictionary, corpus, texts, num_topics, passes, minimum_probability, iterations, decay, alpha):
    coherence_c_v_topic = []
    coherence_u_mass_topic = []
    model_list_topic = []
    log_list = []
    
    index = 0

    for i_topic in num_topics:
      for i_pass in passes:
        for i_prob in minimum_probability:
          for i_it in iterations:
            for i_decay in decay:
              for i_alpha in alpha:
                log_dic = {}
                print(f'Run for index {index}.')
                header_line = f'Log: Run with {i_topic} topics, {i_pass} passes, {i_prob} minimum probability, {i_it} iterations, {i_decay} decay, and {i_alpha} alpha.'
                log_dic['parameters'] = header_line
                print(header_line)
                model = LdaMulticore(corpus=corpus, id2word=dictionary, random_state=267,
                                     num_topics = i_topic,
                                     passes = i_pass,
                                     iterations = i_it,
                                     decay = i_decay,
                                     alpha = i_alpha) 
                model_list_topic.append(model)

                # c v
                coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
                cm_score = coherencemodel.get_coherence()
                cv_line = f'Log: Run ends with a c_v score of {cm_score}.'
                log_dic['c_v'] = cv_line
                print(cv_line)
                coherence_c_v_topic.append(cm_score)

                # u mass
                coherencemodel = CoherenceModel(model=model, corpus=corpus, coherence='u_mass')
                cm_score = coherencemodel.get_coherence()
                u_mass_line = f'Log: Run ends with a u_mass score of {cm_score}.'
                log_dic['u_mass'] = u_mass_line
                print(u_mass_line)
                coherence_u_mass_topic.append(cm_score)
                log_list.append(log_dic)
                index += 1
                print('\n')

    return model_list_topic, coherence_c_v_topic, coherence_u_mass_topic, log_list
```
