# Create Data Bucket for Model Training

## Load relevant packages

```python
# dependencies
import pandas as pd
from nltk.corpus import stopwords
import string
from nltk.stem.wordnet import WordNetLemmatizer
import io
from google.colab import files
from collections import Counter
import nltk
```

## Load Data

```python
# call function to select file to upload
# note: depending on the targeted data set (complete or compact) different input data needs to be loaded
uploaded = files.upload()
```

```python
# initatie empty dataframe for all the content
# load data from CSV into data frame
df_complete = pd.DataFrame()
run = 1

for key, value in uploaded.items():
  print('Log: Run #' + str(run) + ' started.')
  print('Log: File: ' + str(key) + ' is being to read in.')
  # read in data from uploaded csv
  d = pd.read_csv(io.BytesIO(value), dtype=str)
  # check some details per load
  display(d.head(3))
  print(d.columns)
  # add to complete df
  df_complete = df_complete.append(d)
  print('Log: End of run')
  run = run + 1
```
