## Build Data Frame for Survey Question Selection

The following cells build up a dictionary of dataframes in order to make it possible to easily access tweets by category (based on the threshold).

```python
g_thresholds = [0.8, 0.3]
l_thresholds = [0.01, 0.02]
```

```python
# LDA build - only execute for LDA case
dic_of_dic_df = {}

for col in df.columns:
  if col.isnumeric():
    dic_of_df = {}
    prev_thresh = 1
    topic = 'topic ' + str(col)
    df1 = df[df['top_topic'] == col] # ensure only the top values are chosen
    for thresh in g_thresholds:
      threshold = 'greater threshold ' + str(thresh)
      df_tmp = df1[(df1[col] < prev_thresh) & (df1[col] >= thresh)]
      dic_of_df[threshold] = df_tmp
      prev_thresh = thresh
    prev_thresh = 0
    for thresh in l_thresholds:
      topic = 'topic ' + str(col)
      threshold = 'lower threshold ' + str(thresh)
      df_tmp = df[(df[col] < thresh) & (df[col] >= prev_thresh)]
      print(thresh)
      print(len(df_tmp))
      dic_of_df[threshold] = df_tmp
      prev_thresh = thresh
    dic_of_dic_df[topic] = dic_of_df
```

```python
# BERT build - only execute for BERTopic case
dic_of_dic_df = {}

for col in df.columns:
  if col.isnumeric():
    dic_of_df = {}
    prev_thresh = 1
    topic = 'topic ' + str(col)
    df1 = df[df['top topic'] == int(col)] # ensure only the top values are chosen
    for thresh in g_thresholds:
      threshold = 'greater threshold ' + str(thresh)
      df_tmp = df1[(df1[col] < prev_thresh) & (df1[col] >= thresh)]
      dic_of_df[threshold] = df_tmp
      prev_thresh = thresh
    prev_thresh = 0
    for thresh in l_thresholds:
      topic = 'topic ' + str(col)
      threshold = 'lower threshold ' + str(thresh)
      df_tmp = df[(df[col] < thresh) & (df[col] >= prev_thresh)]
      print(thresh)
      print(len(df_tmp))
      dic_of_df[threshold] = df_tmp
      prev_thresh = thresh
    dic_of_dic_df[topic] = dic_of_df
```

## Build Survey

Note: In order to generate multiple surveys the following cells are executed multiple times. Every generation leads to a new unique combination of survey as it works via a random selection and no set.seed function or similar is used. 

```python
test_col = [col for col in df.columns]
test_col.append('group')
test_col
```

```python
# build pairs 
select_topics = [3, 5, 13, 15, 18]
select_g_thresholds = [0.3]
select_l_thresholds = [0.02]
sample_no = 10

# req_col = [col for col in df.columns]
# req_col.append('group')
# df_survey_final = pd.DataFrame(columns=req_col)

df_survey = pd.DataFrame()


group = 1
for st in select_topics:
  first_key = 'topic ' + str(st)
  print(first_key)
  l_group = group
  for sgt in select_g_thresholds:
    second_key = 'greater threshold ' + str(sgt)
    df_tmp = dic_of_dic_df[first_key][second_key]
    df_sample = df_tmp.sample(n = sample_no)
    group_end = group + sample_no
    df_sample['group'] = list(range(group,group_end))
    df_survey = df_survey.append(df_sample)
    group = group_end
  for slt in select_l_thresholds:
    second_key = 'lower threshold ' + str(slt)
    df_tmp = dic_of_dic_df[first_key][second_key]
    df_sample = df_tmp.sample(n = sample_no)
    l_group_end = l_group + sample_no
    df_sample['group'] = list(range(l_group,l_group_end))
    df_survey = df_survey.append(df_sample)
    l_group = l_group_end

df_survey
```

```python
df_shuffle_survey = df_survey.sample(frac=1)
df_shuffle_survey
```

## Store Survey

In the following cells the generated survey is stored. 

```python
df_shuffle_survey = df_shuffle_survey.sort_values(by='group')
df_shuffle_survey
```
