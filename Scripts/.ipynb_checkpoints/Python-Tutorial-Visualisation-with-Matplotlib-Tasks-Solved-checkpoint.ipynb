## Task 1.1
- Load the dataset with pandas and familiarize yourself with it by looking at the head of the data. 

```python
# TODO: Load the dataset and print the head of the data 
import pandas as pd
playstore = pd.read_csv('googleplaystore.csv')
playstore.head()
playstore.sample(n=30)
```

## Task 1.2
Usually the datasets are not prepaired for plotting e.g. missing values / no correct datatypes and so on.
Therefore we will process the dataste a bit more to be usefull:
- clean the dataset / handle null values:
  1. check the number of null values with [`isnull().sum()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isnull.html)
  2. set rating to 0 if missing ([`fillna`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html) function) - apply it only to the rating column
  3. drop all **rows** with remaining null/nan values in any column ([`dropna`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html) function)
- have a look at the types with ([`.dtypes`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dtypes.html)) and change the column `Reviews` to numeric with [`pd.to_numeric`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html)

```python
# TODO: 1. check the number of null values
playstore.isnull().sum()
```

```python
# TODO: 2. set rating to 0 if missing - you can reassign the column to itself and check with isnull().sum() if it worked
playstore["Rating"] = playstore["Rating"].fillna(0)
playstore.isnull().sum()
```

```python
# TODO: 3. drop rows (defaults are good here drop rows and not columns (axis=0) if there is at least one NA(how='any') )
#       check again if it worked out
playstore = playstore.dropna() 
playstore.isnull().sum()
```

```python
#TODO: print the dtypes and change the column Reviews to numeric
print(playstore.dtypes)
playstore['Reviews'] = pd.to_numeric(playstore['Reviews'])
```

## Task 1.3
Plot a bar plot of the content rating to see the distribution of it.

Hints:
- first find out how to get the number of occurances for each content rating
- you can adjust the rotation of the ticks with [`plt.xticks(rotation=60)`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xticks.html)


```python
# TODO: Plot a bar plot of the content rating to see the distribution of it. (don't forget to import matplotlib first)
import matplotlib.pyplot as plt

content_rating = playstore['Content Rating'].value_counts() # use the value_counts function here
plt.bar(content_rating.index, content_rating.values)
plt.xticks(rotation=60)
plt.show()
```

## Task 1.4
- add x and y labels to your plot as well as a title

```python
#TODO: create the same plot as before but add some labels to it
plt.bar(content_rating.index, content_rating.values)
plt.xlabel('Content Rating')
plt.ylabel('Number of Apps')
plt.title('Playstore Apps')
plt.xticks(rotation=60)
plt.show()
```

## Task 2

For the this task, we first want to analyze the price and rating in a scatter plot to see if higher prices results in higher ratings.

To do this, we first need to preprocess the price column because it is not a numeric variable (`$`in front of the values makes it a string). Thus processig is an important step when analyzing data.

The idea is to use the string replace function ([`.str.replace(old, new)`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.replace.html) similar to the string split function from the pandas tutorial. If the character `$` is replaced by an empty string `''` then the column should be parseable ([`pd.to_numeric`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html)).

```python
# TODO: preprocess the price column
playstore['Price'] = playstore['Price'].str.replace('$', '')
playstore['Price'] = pd.to_numeric(playstore['Price'])
```

## Task 2.1
- create a subset of the data where the price is greater than ten
- create a scatter plot with price as x axis and rating as y axis

```python
# TODO: subset of the data where the price is greater than ten 
playstore_costly = playstore[playstore['Price'] > 10]
playstore_costly
```

```python
# TODO: scatter plot with price as x axis and rating as y axis
plt.scatter(playstore_costly['Price'], playstore_costly['Rating'])
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
# TODO: lets have a look at apps with higher prices (> 350)
playstore_very_costly = playstore[playstore['Price'] > 350]
playstore_very_costly
# usually 'I am rich' apps 
```

## Task 2.2
Let us further analze the dataset by looking at the review vs rating scatter plot.
For this we only use the most installed apps (`Installs == '1,000,000,000+'`).  

```python
# TODO: create the subset where Installs == '1,000,000,000+'
playstore_most_installed = playstore[playstore['Installs'] == '1,000,000,000+']
playstore_most_installed.head()
```

```python
plt.scatter(playstore_most_installed['Reviews'], playstore_most_installed['Rating'])
plt.xlabel('Reviews')
plt.ylabel('Rating')
plt.show()
```

- now iterate over the rows and create an annotation for each datapoint which shows the genre

```python
# TODO: scatter plot (Reviews, Rating) with annotations
plt.scatter(playstore_most_installed['Reviews'], playstore_most_installed['Rating'])
for row_index, row in playstore_most_installed.iterrows():
     plt.annotate(row['Genres'],(row['Reviews'] + 0.05, row['Rating']))
plt.xlabel('Reviews')
plt.ylabel('Rating')
plt.show()
```

## Task 2.3
We see that some apps have a high number of reviews. We would like to know in which category these apps are.

The idea is to create an overlay plot where we iterate over the category groups and draw a scatter plot for each group.

```python
# TODO: create a scatter plot like in task 2.2 but now plot the category as color (see Slide Plotting groups)
for name, group in playstore_most_installed.groupby('Category'):
    plt.scatter(group['Reviews'], group['Rating'], label=name)
plt.show()
```

## Task 2.4
Now we see some colors but don't know which color corresponds to which category.
To see the mapping, create a legend and place it on the lower right.

Furthermore we want to increase the figure size ( to 8x8 inches) to see more.

```python
plt.figure(figsize=(8,8))
for name, group in playstore_most_installed.groupby('Category'):
    plt.scatter(group['Reviews'], group['Rating'], label=name)
plt.legend(loc='lower right')
plt.show()
```

- furthermore enable the grid...

```python
# TODO: enable grid 
plt.figure(figsize=(8,8))
for name, group in playstore_most_installed.groupby('Category'):
    plt.scatter(group['Reviews'], group['Rating'], label=name)
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
